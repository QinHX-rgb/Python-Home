```python
# 1 -- 列
# 0 -- 行

df = df.drop("商品图片链接",axis = 1)   # 删除“商品图片链接”列 
```

```python
df
```

```python
df = df.dropna(subset=["销量"])  # 删除含有缺失值的行
```

```python
df
```

```python
df["价格"].mean()  # 计算“价格”列的平均值
```

```python
df["价格"] = df["价格"].fillna(df["价格"].mean())  # 用平均值填充“价格”列的缺失值
```

```python
df
```

```python
df["省/直辖市"] = df["省/直辖市"].fillna("未知")  # 用“未知”填充“省/直辖市”列的缺失值
```

```python
df["市"] = df["市"].fillna("未知")  # 用“未知”填充“市”列的缺失值
```

```python
df.duplicated()  # 检查重复数据
```

```python
df = df.drop_duplicates()  # 删除重复数据
```

```python
df.to_csv("cleaned_taobao_data.csv", index=False)  # 将清洗后的数据保存为CSV文件
```
